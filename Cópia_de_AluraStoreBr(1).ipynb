### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
import matplotlib.pyplot as plt
```

```python
faturamento_loja1 = round(sum(loja['Preço']),2)
faturamento_loja2 = round(sum(loja2['Preço']),2)
faturamento_loja3 = round(sum(loja3['Preço']),2)
faturamento_loja4 = round(sum(loja4['Preço']),2)
faturamento_das_lojas = [faturamento_loja1,faturamento_loja2, faturamento_loja3,faturamento_loja4]
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

for i in range(len(lojas)):
  print(f'O faturamento da loja {lojas[i]} foi de R$ {faturamento_das_lojas[i]}')


```

```python
plt.pie(faturamento_das_lojas, labels=lojas, autopct='%1.1f%%', startangle=90, colors=['gold', 'skyblue', 'lightgreen', 'salmon'])
plt.title('Comparação de faturamento das lojas')

plt.show()
```

# 2. Vendas por Categoria


```python
# Dicionário contendo as categorias e os valores da loja 1
categorias_loja1 = loja['Categoria do Produto'].value_counts().to_dict()

# Variável que armazena a chave da categoria mais popular da loja 1
categoria_mais_popular_loja1 = max(categorias_loja1, key=categorias_loja1.get)

# Dicionário contendo as categorias e os valores da loja 2
categorias_loja2 = loja2['Categoria do Produto'].value_counts().to_dict()

# Variável que armazena a chave da categoria mais popular da loja 2
categoria_mais_popular_loja2 = max(categorias_loja2, key=categorias_loja2.get)

# Dicionário contendo as categorias e os valores da loja 3
categorias_loja3 = loja3['Categoria do Produto'].value_counts().to_dict()

# Variável que armazena a chave da categoria mais popular da loja 3
categoria_mais_popular_loja3 = max(categorias_loja3, key=categorias_loja3.get)

# Dicionário contendo as categorias e os valores da loja 4
categorias_loja4 = loja4['Categoria do Produto'].value_counts().to_dict()

# Variável que armazena a chave da categoria mais popular da loja 4
categoria_mais_popular_loja4 = max(categorias_loja4, key=categorias_loja4.get)

categorias_mais_populares = [categoria_mais_popular_loja1,categoria_mais_popular_loja2,categoria_mais_popular_loja3,categoria_mais_popular_loja4]
```

```python
for i in range(len(categorias_mais_populares)):
  print(f'A categoria mais popular na {lojas[i]} foi *** {categorias_mais_populares[i]} ***')
```

# 3. Média de Avaliação das Lojas

```python
def calcular_media(notas):
  media = sum(notas)/len(notas)
  return round(media,2)

media_avaliacao_loja1 = calcular_media(loja['Avaliação da compra'])
media_avaliacao_loja2 = calcular_media(loja2['Avaliação da compra'])
media_avaliacao_loja3 = calcular_media(loja3['Avaliação da compra'])
media_avaliacao_loja4 = calcular_media(loja4['Avaliação da compra'])
```

```python
medias_de_avaliação = [media_avaliacao_loja1,media_avaliacao_loja2,media_avaliacao_loja3,media_avaliacao_loja4]
cores = ['red', 'blue', 'green', 'orange']

plt.barh(lojas, medias_de_avaliação,color=cores)

plt.xlabel('Media de avaliação')
plt.ylabel('Loja')
plt.title('Média de avaliação das lojas')

plt.style.use('tableau-colorblind10')

for i, valor in enumerate(medias_de_avaliação):
  plt.text(valor -0.1, i, f'{valor}', va='center', ha='right', color='white')

plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
# Loja 1
produtos_loja1 = loja['Produto'].value_counts().to_dict()

produtos_mais_vendidos_loja1 = {
    chave: valor
    for chave,valor in produtos_loja1.items()
    if valor == max(produtos_loja1.values())
}

produtos_menos_vendidos_loja1 = {
    chave:valor
    for chave,valor in produtos_loja1.items()
    if valor == min(produtos_loja1.values())
}


# Loja 2
produtos_loja2 = loja2['Produto'].value_counts().to_dict()

produtos_mais_vendidos_loja2 = {
    chave:valor
    for chave,valor in produtos_loja2.items()
    if valor == max(produtos_loja2.values())
}

produtos_menos_vendidos_loja2 = {
    chave:valor
    for chave,valor in produtos_loja2.items()
    if valor == min(produtos_loja2.values())
}


# Loja 3
produtos_loja3 = loja3['Produto'].value_counts().to_dict()

produtos_mais_vendidos_loja3 = {
    chave:valor
    for chave,valor in produtos_loja3.items()
    if valor == max(produtos_loja3.values())
}

produtos_menos_vendidos_loja3 = {
    chave:valor
    for chave,valor in produtos_loja3.items()
    if valor == min(produtos_loja3.values())
}


# Loja 4
produtos_loja4 = loja4['Produto'].value_counts().to_dict()

produtos_mais_vendidos_loja4 = {
    chave:valor
    for chave,valor in produtos_loja4.items()
    if valor == max(produtos_loja4.values())
}

produtos_menos_vendidos_loja4 = {
    chave:valor
    for chave,valor in produtos_loja4.items()
    if valor == min(produtos_loja4.values())
}

```

```python
produtos_mais_vendidos = [produtos_mais_vendidos_loja1,produtos_mais_vendidos_loja2,produtos_mais_vendidos_loja3,produtos_mais_vendidos_loja4]
for i in range(len(produtos_mais_vendidos)):
  print(f'O(s) produto(s) mais vendidos na {lojas[i]} => {produtos_mais_vendidos[i]}')
```

# 5. Frete Médio por Loja


```python
# Função para calcular o valor médio dos fretes
def calcula_frete(fretes):
  media_fretes = round(sum(fretes) / len(fretes),2)
  return media_fretes

# Loja 1
fretes_loja1 = calcula_frete(loja['Frete'])

# Loja 2
fretes_loja2 = calcula_frete(loja2['Frete'])

# Loja 3
fretes_loja3 = calcula_frete(loja3['Frete'])

# Loja 4
fretes_loja4 = calcula_frete(loja4['Frete'])
```

```python
fretes = [fretes_loja1,fretes_loja2,fretes_loja3,fretes_loja4]
print(fretes)
cores_fretes = ['gold', 'skyblue', 'lightgreen', 'salmon']
from matplotlib.ticker import FuncFormatter

plt.bar(lojas,fretes,color=cores_fretes)
plt.xlabel('Loja')
plt.ylabel('Valor')
plt.title('Frete médio')

def formatar_moeda(valor, pos):
    return f'R${valor:,.2f}'

plt.gca().yaxis.set_major_formatter(FuncFormatter(formatar_moeda))

plt.style.use('_classic_test_patch')
plt.savefig('grafico.png')

plt.show()
```
